```python
from pandas import *
from google.colab import drive
drive.mount('/content/drive')
df = read_csv('/content/drive/MyDrive/data.csv')
print("Original Database")
print(df)

print('\n\n\n')

#prints out null value table to see if data has null valeus
print("Null value table")
res = df.notnull()
print(res)

#variables to store the mnean of each column
duration_mean = df['Duration'].mean()
pulse_mean = df['Pulse'].mean()
maxPulse_mean = df['Maxpulse'].mean()
calories_mean = df['Calories'].mean()

#replaces the null values in each column with their mean, if null values are present
df['Duration'].fillna(value=duration_mean, inplace=True)
df['Pulse'].fillna(value=pulse_mean, inplace=True)
df['Maxpulse'].fillna(value=maxPulse_mean, inplace=True)
df['Calories'].fillna(value=calories_mean, inplace=True)
print(df)

print('\n\n\n')

#aggregating from pulse and calories
print("Aggregated pulse and calories columnn: ")
print(df.aggregate({"Pulse" : ['min', 'max', 'count', 'mean'],
              "Calories":['min', 'max', 'count', 'mean']}))

print('\n\n\n')

#printing out the Calories values and calories and pulse values
print("Values greater than 500 and less than 1000")
print(df.loc[(df['Calories'] > 500) & (df['Calories'] < 1000)])
print('\n\n\n')
print("Calorie Values greater than 500 and pulse values less than 100")
print(df.loc[(df['Calories'] > 500) & (df['Pulse'] < 100)])

print('\n\n\n')

#create new datbase that is the same, but without maxpulse, and proof that the original database is unchanged
df_modified = df.loc[:, df.columns!='Maxpulse']
print("Modified database")
print(df_modified)
print('\n\n\n')
print("Proof that original database is still the same")
print(df)

print('\n\n\n')

#delte the maxpulse column from the database
df = df.drop('Maxpulse', axis=1)
print("Original database with Maxpulse removed")
print(df)

print('\n\n\n')

#convert datatype of calories to int datatype
df = df.astype({"Calories" : int})
print("Database with Calories converted to integers")
print(df)

print('\n\n\n')

#create scatter plot
print("Scatter plot of Duration and Calories")
df.plot.scatter(x = 'Duration', y = 'Calories')
```
